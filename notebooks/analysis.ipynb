```python
data["Estimated expenses (other farm and off-farm)"] = (
    data[excel_to_actual['CO']] * data[excel_to_actual['CP']] * 8100 +
    data[excel_to_actual['CO']] * data[excel_to_actual['CQ']] * 3950 +
    data[excel_to_actual['CO']] * data[excel_to_actual['CR']] * 3700 +
    data[excel_to_actual['CO']] * (
        (data[excel_to_actual['CT']] * 1000 +
         data[excel_to_actual['CU']] * 500 +
         data[excel_to_actual['CV']] * 500 +
         data[excel_to_actual['CW']] * 500 +
         data[excel_to_actual['CX']] * 500 +
         data[excel_to_actual['CY']] * 500 +
         data[excel_to_actual['CZ']] * 500 +
         data[excel_to_actual['DA']] * 500 +
         data[excel_to_actual['DB']] * 500 +
         data[excel_to_actual['DC']] * 500 +
         data[excel_to_actual['DD']] * 500
        )
    ) +
    data[excel_to_actual['DK']] * data[excel_to_actual['DL']] * 2400 +
    data[excel_to_actual['DK']] * data[excel_to_actual['DM']] * 3950 +
    data[excel_to_actual['DK']] * data[excel_to_actual['DN']] * 3700 +
    data[excel_to_actual['DK']] * (
        (data[excel_to_actual['DP']] * 500 +
         data[excel_to_actual['DQ']] * 500 +
         data[excel_to_actual['DR']] * 500 +
         data[excel_to_actual['DS']] * 500 +
         data[excel_to_actual['DT']] * 500
        )
    ) +
    (data[excel_to_actual['EM']] * 3 +
     data[excel_to_actual['FB']] * 3 +
     data[excel_to_actual['FC']] * 3
    )
)
```

```python
data[excel_to_actual['DE']] = data[excel_to_actual['DE']].fillna(0)
data[excel_to_actual['DV']] = data[excel_to_actual['DV']].fillna(0)
data[excel_to_actual['DF']] = data[excel_to_actual['DF']].fillna(0)
data[excel_to_actual['DW']] = data[excel_to_actual['DW']].fillna(0)
data[excel_to_actual['ES']] = data[excel_to_actual['ES']].fillna(0)
data[excel_to_actual['ET']] = data[excel_to_actual['ET']].fillna(0)
data[excel_to_actual['EU']] = data[excel_to_actual['EU']].fillna(0)
data[excel_to_actual['EJ']] = data[excel_to_actual['EJ']].fillna(0)
```

```python
data["Estimated income (another f. and off-farm)"] = (
    data[excel_to_actual['DE']] * data[excel_to_actual['DF']] * (1 / 100) * 5000 +
    data[excel_to_actual['DV']] * data[excel_to_actual['DW']] * (1 / 100) * 1500 +
    data[excel_to_actual['ES']] * 3 +
    data[excel_to_actual['ET']] * 3 +
    data[excel_to_actual['EU']] * 3 +
    data[excel_to_actual['EJ']] * 3
)
```

```python
def repayment_status(row):
    if row[excel_to_actual['FH']] == "No":
        return "No Repayment Issues"
    elif row[excel_to_actual['FJ']] == "less than 3 months":
        return "Short-term Delay"
    elif row[excel_to_actual['FJ']] == "less than 6 months":
        return "Moderate Delay"
    elif row[excel_to_actual['FJ']] == "1 year":
        return "Extended Delay"
    else:
        return "Extended Delay"
data["Default + length (HISTORY)"] = data.apply(repayment_status, axis=1)
```

```python
data[excel_to_actual['AI']] = data[excel_to_actual['AI']].fillna('No')
data[excel_to_actual['AK']] = data[excel_to_actual['AK']].fillna('No')
data[excel_to_actual['AM']] = data[excel_to_actual['AM']].fillna('No')
data[excel_to_actual['AN']] = data[excel_to_actual['AN']].fillna('No')
data[excel_to_actual['AO']] = data[excel_to_actual['AO']].fillna('No')
data[excel_to_actual['AP']] = data[excel_to_actual['AP']].fillna('No')
data[excel_to_actual['AQ']] = data[excel_to_actual['AQ']].fillna('No')
```

```python
def evaluate_status(row):
    yes_count = (
        (row[excel_to_actual['AI']] == "Yes") +
        (row[excel_to_actual['AK']] == "Yes") +
        (row[excel_to_actual['AM']] == "Yes") +
        (row[excel_to_actual['AN']] == "Yes") +
        (row[excel_to_actual['AO']] == "Yes") +
        (row[excel_to_actual['AP']] == "Yes") +
        (row[excel_to_actual['AQ']] == "Yes")
    )
    
    if yes_count < 4:
        return "Poor"
    elif yes_count < 6:
        return "Moderate"
    else:
        return "Good"
data["Social Status"] = data.apply(evaluate_status, axis=1)
```

```python
def evaluate_status(row):
    yes_count = (
        (row[excel_to_actual['FM']] == "Yes") +
        (row[excel_to_actual['FK']] == "Yes") +
        (row[excel_to_actual['FG']] == "Yes") +
        (row[excel_to_actual['FE']] == "Yes") +
        (row[excel_to_actual['FD']] == "Yes") +
        (row[excel_to_actual['EK']] == "Yes")
    )
    
    if yes_count <= 2:
        return "Poor"
    elif yes_count <= 4:
        return "Moderate"
    else:
        return "Good"

data["Credit History  (proxy )"] = data.apply(evaluate_status, axis=1)
```

```python
final_data = pd.read_excel("../data/output_custom_order.xlsx")
num_cols = len(final_data.columns)
excel_headers = list(string.ascii_uppercase)
if num_cols > 26:
    excel_headers += [f"{char1}{char2}" for char1 in string.ascii_uppercase for char2 in string.ascii_uppercase]
excel_to_actual = {excel_headers[i]: col_name for i, col_name in enumerate(final_data.columns)}
```

```python
def calculate_score(row):
    score = 0 if row[excel_to_actual['B']] == "Male" else 5
    
    age = row[excel_to_actual['C']]
    if age <= 25:
        score += 0
    elif age <= 30:
        score += 2.5
    elif age <= 38:
        score += 5
    elif age <= 59:
        score += 10
    else:
        score += 0
    
    employment = row[excel_to_actual['D']]
    if employment == 0:
        score += 5
    elif 1 <= employment < 4:
        score += 3
    elif 4 <= employment < 5:
        score += 1
    else:
        score += 0

    education = row[excel_to_actual['E']]
    if education <= 0:
        score += 0
    elif 0 < education < 1:
        score += 2
    elif 1 <= education < 2:
        score += 3
    elif 2 <= education < 3:
        score += 4
    else:
        score += 5
    
    score += 7.5 if row[excel_to_actual['F']] == "Yes" else 0
    
    financial = row[excel_to_actual['G']]
    if financial <= 0.5:
        score += 0
    elif financial < 1:
        score += 5
    elif financial < 2:
        score += 10
    else:
        score += 15
    
    living_situation = row[excel_to_actual['H']]
    if living_situation == "Traditional all-in-one":
        score += 0
    elif living_situation == "Traditional with multiple rooms and shelters":
        score += 5
    elif living_situation == "Modern":
        score += 10
    else:
        score += 0
    
    income = row[excel_to_actual['I']]
    if income < 21300:
        score += 15
    elif income < 27900:
        score += 10
    elif income < 49300:
        score += 5
    else:
        score += 0
    
    additional_income = row[excel_to_actual['J']]
    if additional_income < 37500:
        score += 0
    elif additional_income < 90000:
        score += 5
    elif additional_income < 114478:
        score += 10
    else:
        score += 15
    

    savings = row[excel_to_actual['K']]
    if savings < 1000:
        score += 10
    elif savings < 5000:
        score += 7
    elif savings < 11000:
        score += 4
    else:
        score += 0
    
  
    loan = row[excel_to_actual['L']]
    if loan < 2000:
        score += 0
    elif loan < 6000:
        score += 4
    elif loan < 25000:
        score += 7
    else:
        score += 10
    
  
    repayment = row[excel_to_actual['M']]
    if repayment == "No Repayment Issues":
        score += 10
    elif repayment == "Short-term Delay":
        score += 7
    elif repayment == "Moderate Delay":
        score += 4
    else:
        score += 0
   
    loan_amount = row[excel_to_actual['N']]
    if loan_amount == 0:
        score += 3.75
    elif loan_amount <= 10000:
        score += 2
    else:
        score += 0
    
    status = row[excel_to_actual['O']]
    if status == "Good":
        score += 10
    elif status == "Moderate":
        score += 5
    else:
        score += 0
    
    
    storage = row[excel_to_actual['P']]
    score += 5 if storage == "Aggregator storage nearby" else 0
    
  
    farming = row[excel_to_actual['Q']]
    if farming == "Modern (mechanized farming) owned":
        score += 5
    elif farming == "Traditional (oxen plowing) owned":
        score += 4
    elif farming == "Traditional (oxen plowing) owned & Modern (mechanized farming) rented":
        score += 3
    elif farming == "Modern (mechanized farming) rented":
        score += 1
    else:
        score += 0
    
 
    score += 5 if row[excel_to_actual['R']] == "Yes" else 0
    

    final_status = row[excel_to_actual['S']]
    if final_status == "Poor":
        score += 0
    elif final_status == "Moderate":
        score += 2
    else:
        score += 3.73 
    return score

final_data["score_mb"] = final_data.apply(calculate_score, axis=1)
```

```python
def percentrank_inc(final_data, value):
    return percentileofscore(final_data, value, kind='rank') / 100
final_data["PERCENTILE"] = final_data['score_mb'].apply(lambda x: percentrank_inc(final_data['score_mb'], x) * 100)
```

```python
def eligibility_status(value):
    if value < 60:
        return "Not Eligible"
    elif value < 70:
        return "Review"
    else:
        return "Eligible"
final_data["DECITION"] = final_data["score_mb"].apply(eligibility_status)
```
